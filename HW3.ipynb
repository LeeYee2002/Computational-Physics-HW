**(1pt) The variance is the second central moment, $σ^2 = \langle (x−µ)^2 \rangle$ . Prove it is equal to $x^2 −{⟨x⟩}^2$**

$$\langle (x−µ)^2 \rangle = \langle \sum_{i}^n (x_i-\mu)^2 \rangle$$

Where $\mu$ is the average of samples: $$\mu = \frac{\sum^n_{i} x_i}{n}$$

Expand the square:

$$\langle \sum_{i}^n (x_i-\mu)^2 \rangle = \langle \sum_{i}^n x_i^2 -2\mu x_i + \mu^2 \rangle$$

From the above relation:

$$\langle \sum_{i}^n x_i^2 -2\mu x_i + \mu^2 \rangle = \langle \sum_{i}^nx_i^2 -2\mu \sum_{i}^n x_i + \sum_{i}^n \mu^2 \rangle = \langle \sum_{i}^nx_i^2 -2\mu (n \mu) + n \mu^2 \rangle = \langle x \rangle$$


Rearrange the equation: 
$$\langle \sum_{i}^nx_i^2 -2\mu (n \mu) + n \mu^2 \rangle = \langle \sum_{i}^nx_i^2 -n\mu^2  \rangle$$

Use the sumation notation for $\mu$:
$$\langle \sum_{i}^nx_i^2 -\sum_{i}^n \mu^2  \rangle$$
We know the expectation value have simple additivity:

$$\langle \sum_{i}^nx_i^2 -\sum_{i}^n \mu^2  \rangle = \langle \sum_{i}^nx_i^2\rangle -\langle \sum_{i}^n \mu^2  \rangle = \langle \sum_{i}^nx_i^2\rangle - \mu^2 =\langle x^2\rangle - \langle x \rangle ^2$$

Such that:

$$σ^2 = \langle (x−µ)^2 \rangle =\langle x^2\rangle - \langle x \rangle ^2$$

**(3pt) To third order G(k) = $\exp(ikµ − \frac{k^2}{2} σ^2 − \frac{ik^3}{6} κ_3 +O(k^4) )$. Prove that this third cumulant κ_3 is the third central moment $(x−µ)^3$**:

Expand $G(k)$ to third order:

$$G(k) = 1 +ik\mu -\frac{1}{2} k^2 \langle x^2 \rangle -\frac{ik^3}{6} \langle x^3 \rangle + O(k^4)...$$

We first analyze $\kappa_3$:
$$\kappa_3 = \langle (x-\mu)^3 \rangle = \langle x^3-3x^2\mu +3x \mu^2 -\mu^3 \rangle = \langle x^3 \rangle -3\mu(\langle x^2 \rangle - \mu \langle x \rangle) -\mu^3$$
Then we focus on the latter part:
$$ \langle x^3 \rangle -3\mu(\langle x^2 \rangle - \mu \langle x \rangle) -\mu^3 = \langle x^3 \rangle -3\mu(\langle x^2 \rangle - \mu^2) -\mu^3 $$


Where we see the variance; thus, we can write $\kappa_3$ as:

One can rearrange the formula:

$$\langle x^3 \rangle = \kappa_3 +3\mu \sigma^2 +\mu^3$$

Return to the expansion of $G(k)$, we insert the relation above, where $G(k)$ becomes:

$$G(k) = 1 +ik\mu -\frac{1}{2}k^2\langle x^2 \rangle -\frac{ik^3}{6} (\kappa_3 +3\mu \sigma^2 +\mu^3) + O(k^4)...$$

$$G(k) = \exp(ik\mu -\frac{1}{2} k^2\sigma^2 -\frac{ik^3}{6}\kappa_3 +O(k^4)) = \exp(ik \mu) \cdot \exp(-\frac{1}{2} k^2\sigma^2) \cdot \exp(-\frac{ik^3}{6}\kappa_3) \cdot \exp(O(k^4))$$

**The Laplace distribution is the two-sided exponential distribution**
 
 
$$L_{µ,b}(x) = \frac{1}{2b}\exp(\frac{−|x−µ|}{b})$$

**We can see obviously the distribution is symmetric around µ(so µ is the median of $L_{µ,b}$), and the probability peaks there (so µ is the mode of $L_{µ,b}$).**

**Perform X = 1000 experiments where you draw T = $2^{11}$ iid samples from $L_{0,1}$. Let’s imagine the samples in any experiment are drawn sequentially and we can say we accumulate more and more samples as a function of time $t$.**

**• (1pt) For a single experiment make a histogram of all the samples x and compare it to $L_{0,1}$**

Such $L_{0,1}$:

$$L_{0,1}(x) = \frac{1}{2} \exp(- |x| )$$

```python
def laplace(mu,b,x):
    import math
    e =math.e
    return 1/(2*b)*e**(-abs(x-mu)/b)
```

```python
import numpy as np
import matplotlib.pyplot as plt
mu, b = 0, 1
experiments = 1
samples = 2**11
result = np.random.laplace(mu,b, samples)

x_0 = np.linspace(-10,10,1000) #as we see laplace distribution in this case most falls with -10~10
y_0 = laplace(mu,b,x_0)
plt.hist(result, bins = 100, label = '$L_{0,1}$ histogram', density= True)
plt.plot(x_0,y_0, color ='r', label ='$L_{0,1}$ function')
plt.title("Single Laplace Experiment")
plt.legend()
plt.show()
```
